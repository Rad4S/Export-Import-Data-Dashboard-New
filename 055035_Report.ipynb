- **Import/Export Status (Categorical Variable)**:

Understanding how frequently each method is used can help businesses optimize their logistics, identify faster or more affordable routes, and make informed decisions when negotiating with logistics providers.

- **Payment Terms (Categorical Variable)**:  
  - This variable defines the **terms of payment** for each transaction, such as pre-paid, post-paid, or credit-based terms. Payment terms directly impact a company's cash flow and financial management. Longer credit periods may affect liquidity, while advance payments lower financial risk. By analyzing the payment terms for various transactions, businesses can optimize their financial strategies. For example, frequent use of credit terms might require tighter financial controls, while prepaid terms provide immediate cash flow benefits.

- **Non-Categorical Variables**:
  - Variables like **Quantity, Value, Weight**, and **Customs_Code** are continuous and numerical, representing measurable aspects oand financial planning.

## Analysis

1. **Quantity vs. Value (Scatter Plot)**  
   **Observation:** The scatter plot shows no clear trend or correlation between the quantity of goods traded and their value. The data points appear uniformly distributed across the plot, indicating that the relationship between quantity and value is not linear or easily predictable from the visual.  
   **Insight:** Since there is no observable relationship between quantity and value, it suggests that the value of goods doesn't depend on their quantity. This could indicate that factors other than quantity, such as the type of goods or external market conditions, are driving the value. Management strategies should focus on these external variables rather than purely on the volume of goods traded.

2. **Percentage of High-Value Transactions (Pie Chart)**  
   **Observation:** A small percentage of transactions account for the majority of the trade value, with 90% of transactions being lower in value.  
   **Insight:** The business should focus on nurturing and securing these high-value transactions, as they have a disproportionately large impact on revenue.

3. **Frequency of Shipping Methods (Bar Plot)**  
   **Observation:** Shipping via air is used far more frequently than other shipping methods like shipping via land and sea. This can reflect customer preferences, cost efficiency, or logistical constraints.  
   **Insight:** By analyzing the cost and efficiency of the most-used shipping methods, the business can optimize logistics and potentially reduce costs.

4. **Transaction Value Distribution (Histogram)**  
   **Observation:** The distribution is skewed towards lower-value transactions, with a long tail representing the high-value transactions. The KDE (Kernel Density Estimation) curve further highlights the concentration of transactions in the lower range.  
   **Insight:** While a majority of transactions are of lower value, the business should explore ways to increase the average transaction value through bundled offers or upselling strategies.

5. **Weight Distribution per Product Category (Box-Whisker Plot)**  
   **Observation:** There is significant variance in the weight distribution across different product categories. Some categories like Machinery have higher weight ranges, while others like clothing tend to be lighter.  
   **Insight:** Product-specific logistics strategies should be developed. For heavier products, efficient bulk transportation (e.g., sea freight) could be prioritized, while more expensive, lighter products might benefit from faster shipping options.

6. **Monthly Transaction Trends (Line Plot)**  
   **Observation:** There are discernible peaks in certain months, indicating a seasonal pattern in trade. For example, trade volumes spike during certain months, possibly due to holidays or demand cycles.  
   **Insight:** Understanding this seasonality allows for better inventory management, staffing, and financial planning. Strategic campaigns or promotions can also be timed to coincide with these high-demand periods.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Loading the Data.
data = pd.read_csv(r"C:\Users\Radhika\Downloads\Imports_Exports_Dataset.csv")
data.head()
```

```python
# Dimensions of Data.
data.shape
```

```python
# Creating an Unique Sample of 3001 Records using Student Roll Number as Random State.
sample=data.sample(n=3001, random_state=55035)
```

```python
# Dimensions of Sample Data.
sample.shape
```

```python
# Ensure plots are displayed within the notebook
%matplotlib notebook  # Change to interactive plotting mode
```

```python
# 1. Scatter plot for Quantity vs. Value
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Quantity', y='Value', data=sample, color='red')  # Change 'red' to any color you prefer
plt.title('Scatter Plot of Quantity vs. Value')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.show()
```

```python
# 2. Percentage of High-Value Transactions
high_value_transactions = sample[sample['Value'] >= sample['Value'].quantile(0.90)]
percentage_high_value = len(high_value_transactions) / len(sample) * 100

# Plotting the Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(
    [percentage_high_value, 100 - percentage_high_value],
    labels=['High Value', 'Others'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['#800080', '#FFC0CB']  # Purple and Pink colors
)
plt.title('Percentage of High-Value Transactions')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 3. Bar plot for Frequency of Shipping Methods
plt.figure(figsize=(8, 6))
sns.countplot(x='Shipping_Method', data=sample, color='brown')  # Change color to brown
plt.title('Bar Plot of Shipping Methods')
plt.xlabel('Shipping Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Transaction Value Distribution (Histogram)
plt.figure(figsize=(10, 6))

# Plotting histogram with orange color
sns.histplot(sample['Value'], bins=30, kde=True, color='yellow', alpha=0.7)  # Histogram with orange

# Overlay with a slightly transparent yellow color
sns.histplot(sample['Value'], bins=30, kde=True, color='orange', alpha=0.3)  # Overlay with yellow

plt.title('Transaction Value Distribution')
plt.xlabel('Transaction Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Weight Distribution per Product Category (Box-Whisker)
plt.figure(figsize=(12, 6))

# Setting a custom color for the boxplot
sns.boxplot(x='Category', y='Weight', data=sample, palette='pastel')  # Using pastel colors for aesthetics

plt.title('Weight Distribution per Product Category')
plt.xlabel('Category')
plt.ylabel('Weight')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Monthly Transaction Trends (Line)
# Ensure the 'Date' column is in the correct format
sample['Date'] = pd.to_datetime(sample['Date'], format='%d-%m-%Y', dayfirst=True)

# Extract month and create a new column
sample['Month'] = sample['Date'].dt.month

# Group by month and calculate total value
monthly_transactions = sample.groupby('Month')['Value'].sum().reset_index()

# Plotting the Monthly Transaction Trends
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Value', data=monthly_transactions, color='#D8BFD8', linewidth=2)  # Lilac color
plt.title('Monthly Transaction Trends')
plt.xlabel('Month')
plt.ylabel('Total Value')
plt.xticks(rotation=0)
plt.show()
```
